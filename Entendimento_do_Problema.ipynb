# Entendimento do Problema


## **Descrição**  
  O conjunto de dados escolhido foi o **Heart Disease UCI**, este conjunto foi criado com o intuito de classificar se os pacientes listados no mesmo possuem ou não doenças cardíacas, para isso temos alguns dados que podem ser relevantes para realizar a classificação. 
##  **Objetivo**
 Meu objetivo com esse conjunto de dados é criar um modelo que possa predizer se determinado paciente terá ou não alguma doença cardíaca.

## **Variáveis**

*   *age*:Idade;

*   *sex*: Sexo. 0-feminino, 1-masculino;

*   *cp*: Tipo de dor no peito;

*   *trestbps*: Pressão arterial em repouso em mmHg (milímetros de mercurio);

*   *chol*: Colesterol sérico em mg/dl (miligramas por cecilitro);

*   *fbs*: Açucar no sangue em jejum > 120 mg/dl.     0-falso, 1-verdadeiro;

*   *restecg*: Resultados eletrocardiográficos quando em repouso;

*   *thalach*: Frequência cardíaca máxima alcançada;

*   *exang*: Angina induzida por exercício. 0-falso, 1-verdadeiro;

*   *oldpeak*: Depressão ST induzida por exercício em relação ao repouso;

*   *slope*: a inclinação do segmento ST do pico do exercício;

*   *ca*: Número de vasos principais (0-3) coloridos por fluoroscopia;

*   *thal*: 1 = normal; 2 = defeito fixo; 3 = defeito reversível;

*   *target*: Possui a doença ou não. 0-não possui, 1-possui;
